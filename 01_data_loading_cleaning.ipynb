```python
import pandas as pd

file_path = r"C:\Users\pavansai\OneDrive\Desktop\RetailPulse\data\raw\online_retail.xlsx"

df = pd.read_excel(file_path, engine="openpyxl")

df.head()
```

```python
df_clean = df.copy()

# Remove rows where CustomerID is missing
df_clean = df_clean.dropna(subset=["CustomerID"])

# Remove cancelled/returned orders
df_clean = df_clean[~df_clean["InvoiceNo"].astype(str).str.startswith("C")]

# Remove invalid quantity and price
df_clean = df_clean[df_clean["Quantity"] > 0]
df_clean = df_clean[df_clean["UnitPrice"] > 0]

# Convert CustomerID from decimal to integer
df_clean["CustomerID"] = df_clean["CustomerID"].astype(int)

# Create revenue column
df_clean["Revenue"] = df_clean["Quantity"] * df_clean["UnitPrice"]

df_clean.head()
```

## Data Cleaning Summary

Original dataset shape: 541909 rows and 8 columns.

Cleaning steps performed:
1. Removed rows with missing CustomerID.
2. Removed cancelled invoices starting with C.
3. Removed rows where Quantity <= 0.
4. Removed rows where UnitPrice <= 0.
5. Converted CustomerID into integer format.
6. Created Revenue column using Quantity * UnitPrice.
7. Saved cleaned data as cleaned_online_retail.csv.

Final cleaned dataset shape: 397884 rows and 9 columns.
